```
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```
data = load_iris()
data.keys()
```

```
X = data.data
y = data.target
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```
tree = DecisionTreeClassifier(max_depth=14)
tree.fit(X_train,y_train)
```

More options are avalible :
* max_depth
* min_samples_split   -> minimum number of samples a node must have before it can be split
* min_smaples_leaf    -> minimum number of samples a leaf node must have
* max_leaf_nodes ->  maximum number of leaf nodes
* max_features -> maximum number of features that are evaluated for splittin at each node

```
export_graphviz(tree,
               out_file='iris_tree.dot',
               feature_names=data.feature_names,
               class_names = data.target_names,
               rounded = True,
               filled=True)
```
