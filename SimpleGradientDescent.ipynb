```
import numpy as np
import matplotlib.pyplot as plt
from time import sleep

x = 2
l = 0.01
acc = 0.01


while np.abs(2*(x-5)) > acc :
    x = x - l*(2*(x-5))
    sleep(0.1)
    print(x)
    
```
