#Gradient Descent for 1D

```
import numpy as np
import matplotlib.pyplot as plt
from time import sleep

x = 2
l = 0.01
acc = 0.01


while np.abs(2*(x-5)) > acc :
    x = x - l*(2*(x-5))
    sleep(0.1)
    print(x)
    
```

#Gradient Descent for 2D

```
import numpy as np
import matplotlib.pyplot as plt
from time import sleep

x = 0
y = 0
l = 0.01
acc = 0.01


while np.abs(2*(x-5) + 2*(y-3)) > acc :
    x = x - l*(2*(x-5))
    y = y - l*(2*(y-3))
    sleep(0.01)
    print('x={} \t\t y={}'.format(x,y))
    
```
